```python
# Import numpy:
import numpy as np
```

```python
# Initialize a random number generator:
rng = np.random.default_rng(42)
```

```python
# Define a function to check for equality:
def check_equality(a, b, n):
    return ((a == b).sum() == n)
```

```python
# Define a function to check for intersections:
def check_intersections(a, b, n):
    a = set(a)
    b = set(b)
    return (len(a & b) == n)
```

```python
# Create a matrix for constraints:
b = np.array([[6, 8, 2],
              [6, 1, 4],
              [2, 0, 6],
              [7, 3, 8],
              [7, 8, 0]])
```

```python
# Define a list for available numbers to draw from:
numbers = np.arange(10)
```

```python
%%time
# Run 1000 simulations:
n_iters = 0
for i in range(1000):
    
    a = rng.choice(a=numbers, size=3, replace=False)
    score = 0
    
    n_iters += 1
    if check_equality(a, b[0], 1) and check_intersections(a, b[0], 1):
        score += 1
    if check_equality(a, b[1], 0) and check_intersections(a, b[1], 1):
        score += 1
    if check_equality(a, b[2], 0) and check_intersections(a, b[2], 2):
        score += 1
    if check_intersections(a, b[3], 0):
        score += 1
    if check_equality(a, b[4], 0) and check_intersections(a, b[4], 1):
        score += 1
                                                          
    if score == 5:
        print(f"The correct combination: {a}")
        print(f"Number of iterations: {n_iters}")
        break
```
